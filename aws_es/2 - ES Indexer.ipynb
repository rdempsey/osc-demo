```python
# Imports
import json
import pandas as pd
import numpy as np
from espandas import Espandas
import uuid
from datetime import datetime
```

```python
# Load the CSV data
df = pd.read_csv("yts_data_fully_enriched_clean.csv", index_col=False, header=0)
len(df)
```

```python
# Check the data types
df.dtypes
```

```python
# Create an ID column of UUIDs and add it to the dataframe
index_ids = list()
for i in range(0,8610):
    index_ids.append(i)

df['indexId'] = index_ids
```

```python
# Create a datetime column so we have one for Elasticsearch
df['created_at'] = datetime.utcnow().strftime("%Y/%m/%d %H:%M:%S")
df['updated_at'] = datetime.utcnow().strftime("%Y/%m/%d %H:%M:%S")
```

```python
# Let's look at some of the data
df.head()
```

```python
# Index the dataframe into Elasticsearch
INDEX = 'movies'
TYPE = 'movie'
esp = Espandas()
esp.es_write(df, INDEX, TYPE)
```

```python
# Query for the first ten rows and see that they match the original
k = df.indexId[0:10]
res = esp.es_read(k, INDEX, TYPE)
res == df.iloc[0:10].astype('str')
```
